```python
data = df_trans.groupby(['merchant_mcc'], as_index = False).aggregate({'customer_id': 'count','transaction_amt': 'sum'})
```

```python
data = data.sort_values(by='transaction_amt',ascending=False)
data
```

```python
id_kol = len(df_trans.customer_id.unique())
id_kol
```

```python
mcc = list(data.merchant_mcc)
```

```python
for x in mcc:
    a = df_trans.loc[(df_trans.merchant_mcc == x)].customer_id.unique()
    data.loc[data.merchant_mcc == x, '%'] = len(a)*100/id_kol
```

```python
data = data.rename(columns={'customer_id': 'kolvo', 'transaction_amt': 'sum_$'})
data
```

```python
data.to_csv('mcc_pr.csv', index=False)
```

```python
writer = pd.ExcelWriter('mcc_pr.xlsx') 
data.to_excel(writer) 
writer.save()
```
